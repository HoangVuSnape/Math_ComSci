```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
# Gửi request để lấy response
url = "https://diemthi.tuyensinh247.com/diem-chuan/dai-hoc-fpt-FPT.html?success=true&ad_id=5e6ae67c7f8b9a495c8b4567&countdown_info=[object Object]"
response = requests.get(url)
    
# Parse HTML bằng BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Tìm bảng và các hàng
table = soup.find('table')
thead = table.find('thead').find_all('th')
tbody = table.find('tbody').find_all('tr')

# Lấy tiêu đề cột
columns = [header.text.strip() for header in thead]

# Lấy dữ liệu từng hàng
data = []
for row in tbody:
    cols = row.find_all('td')
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Chuyển dữ liệu thành DataFrame
df = pd.DataFrame(data, columns=columns)

# Hiển thị DataFrame
print(df)
```
